# Activation Functions Comparison

This notebook provides an overview and visual comparison of common activation functions used in neural networks: Sigmoid, Tanh, Softmax, ReLU, and Leaky ReLU.

For each function, we'll define its mathematical representation and its derivative, then plot both to understand their behavior. Finally, a comparative plot will be generated.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define Activation Functions and Their Derivatives

```python
def sigmoid(z):
    """Sigmoid activation function."""
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    """Derivative of the Sigmoid function."""
    s = sigmoid(z)
    return s * (1 - s)

def tanh(z):
    """Hyperbolic Tangent (tanh) activation function."""
    return np.tanh(z)

def tanh_derivative(z):
    """Derivative of the Tanh function."""
    return 1 - np.tanh(z)**2

def relu(z):
    """Rectified Linear Unit (ReLU) activation function."""
    return np.maximum(0, z)

def relu_derivative(z):
    """Derivative of the ReLU function."""
    return (z > 0).astype(float) # Returns 1 if z > 0, 0 otherwise

def leaky_relu(z, alpha=0.01):
    """Leaky Rectified Linear Unit (Leaky ReLU) activation function."""
    return np.maximum(alpha * z, z)

def leaky_relu_derivative(z, alpha=0.01):
    """Derivative of the Leaky ReLU function."""
    dz = np.ones_like(z)
    dz[z < 0] = alpha
    return dz

def softmax(z):
    """Softmax activation function."""
    # To prevent overflow, subtract the maximum value from z before exponentiating
    exp_z = np.exp(z - np.max(z, axis=-1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=-1, keepdims=True)
```

Generate Input Values

```python
z = np.linspace(-5, 5, 200) # Increased points for smoother plots
```

Plot Individual Activation Functions and Derivatives

```python
# Sigmoid
plt.figure(figsize=(12, 6))
plt.plot(z, sigmoid(z), label='Sigmoid Activation', color='g')
plt.plot(z, sigmoid_derivative(z), label='Sigmoid Derivative', color='r', linestyle='--')
plt.title('Sigmoid Activation & Derivative')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.grid(True)
plt.legend()
plt.show()

# Tanh
plt.figure(figsize=(12, 6))
plt.plot(z, tanh(z), label='Tanh Activation', color='g')
plt.plot(z, tanh_derivative(z), label='Tanh Derivative', color='r', linestyle='--')
plt.title('Tanh Activation & Derivative')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.grid(True)
plt.legend()
plt.show()

# ReLU
plt.figure(figsize=(12, 6))
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_derivative(z), label='ReLU Derivative', color='r', linestyle='--')
plt.title('ReLU Activation & Derivative')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.grid(True)
plt.legend()
plt.show()

# Leaky ReLU
plt.figure(figsize=(12, 6))
plt.plot(z, leaky_relu(z), label='Leaky ReLU Activation (α=0.01)', color='g')
plt.plot(z, leaky_relu_derivative(z), label='Leaky ReLU Derivative (α=0.01)', color='r', linestyle='--')
plt.title('Leaky ReLU Activation & Derivative')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.grid(True)
plt.legend()
plt.show()
```

## Comparative Plot (Activations Only)
Let's compare the main activation functions on a single plot to see their different output ranges and behaviors.

```python
plt.figure(figsize=(10, 7))

plt.plot(z, sigmoid(z), label='Sigmoid', linestyle='-', linewidth=2)
plt.plot(z, tanh(z), label='Tanh', linestyle='--', linewidth=2)
plt.plot(z, relu(z), label='ReLU', linestyle=':', linewidth=2)
plt.plot(z, leaky_relu(z), label='Leaky ReLU (α=0.01)', linestyle='-.', linewidth=2)

plt.title('Comparison of Activation Functions')
plt.xlabel('Input Value (z)')
plt.ylabel('Output Value')
plt.ylim(-1.5, 1.5) # Set y-limits to better compare range
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
